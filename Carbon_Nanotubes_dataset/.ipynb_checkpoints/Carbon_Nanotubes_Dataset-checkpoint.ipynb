### Outliers

```python
# Finding the Outliers in Chiral_indice_n column:

outlier = sns.boxplot(x="Chiral_indice_n", data=new_data)
```

```python
def outlier(Chiral_indice_n):
    sorted(Chiral_indice_n)
    Q1,Q3 = np.percentile(Chiral_indice_n, [25,75])
    IQR = Q3 - Q1
    lower_range = Q1 - (1.5 * IQR)
    upper_range = Q3 + (1.5 * IQR)
    return lower_range,upper_range

low,up = outlier(new_data.Chiral_indice_n)
print("The lowerbound: ",low)
print("The upperbound: ",up)
```

```python
new_data[(new_data.Chiral_indice_n < low) | (new_data.Chiral_indice_n > up)]
```

```python
new_data.drop(new_data[(new_data.Chiral_indice_n < low) | (new_data.Chiral_indice_n > up)].index, inplace=True)
```

```python
# After removing the outliers in Chiral_indice_n column:

sns.boxplot(x='Chiral_indice_n', data=new_data)
```

### train and test split method

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=12)
```

### Apply the model:

```python
# Linear Algorithm
from sklearn import linear_model 
model = linear_model.LinearRegression()

model.fit(X_train, y_train)
```

#### Now we have trained the model based on X and y
#### So,We are going to predict the test data based upon the trained data

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
# prediction with new value

print(model.predict([[2,1,679005,701318,17033]]))    # (721039,730232,17014 answer)
```

### R2 Score

```python
from sklearn.metrics import r2_score 

print("R2 Score",r2_score(y_test, y_pred))
```

## Accuracy:

#### This method is called train and test split method. Here we can find the accuracy of the data based upon the model that we are training. In the below cell we have used linear regression.

```python
# Linear Regression

from sklearn.metrics import r2_score 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=12)
new_model = linear_model.LinearRegression()
new_model.fit(X_train, y_train) 
y_pred = new_model.predict(X_test)
print("R2 Score",r2_score(y_test, y_pred)) 
```

#### Trying with various regression algorithms and to conclude one best fit algorithm

## Ridge Algorithm

```python
from sklearn.linear_model import Ridge

model_RD = Ridge()
model_RD.fit(X_train, y_train)
y_pred = model_RD.predict(X_test)
print("R2 Score",r2_score(y_test,y_pred))
```

## Lasso Algorithm

```python
from sklearn.linear_model import Lasso 

model_LS = Lasso()
model_LS.fit(X_train, y_train)
y_pred = model_LS.predict(X_test)
print("R2 Score",r2_score(y_test,y_pred))
```

## Elastic Net Regression

```python
from sklearn.linear_model import ElasticNet 

model_EN = ElasticNet()
model_EN.fit(X_train, y_train)
y_pred = model_EN.predict(X_test)
print("R2 Score",r2_score(y_test,y_pred))
```

## Random Forest Regreesion

```python
from sklearn.ensemble import RandomForestRegressor

model_RF = RandomForestRegressor(max_depth=5, random_state=0)
model_RF.fit(X_train, y_train)
y_pred = model_RF.predict(X_test)

print("R2 Score",r2_score(y_test, y_pred))
```

## Conclusion 

## For this Carbon_Nanotubes dataset Random Forest Algorithm is the best fit algorithm as it gives the maximum accuracy of 59%
